```python
from scipy.ndimage import gaussian_filter
```

```python
domain = np.linspace(0, 10, 50)
f = lambda x: 0.1 * x**3 - x**2 + 2
values = f(domain)
noisy = values + np.random.normal(1, 1, 50)
smooth1 = gaussian_filter(noisy, sigma=1)
smooth2 = gaussian_filter(noisy, sigma=3)
scatter = lambda y: Scatter(x=domain, y=y)
py.iplot({
    "data": [scatter(values), 
             Scatter(x=domain, y=noisy),
             Scatter(x=domain, y=smooth1),
             Scatter(x=domain, y=smooth2)],
    "layout": Layout(title="hello world")
})
```

```python
rain_p = np.array([[0.7, 0.3], [0.3, 0.7]])
umbrella_p = np.array([[0.8, 0.1], [0.2, 0.9]])
```

now generate n samples from the distribution
for each sample, choose at random whether it rains or not on the start day
let 0 rain and 1 no rain? then prob dist given by 0/1 column

simple: array of n random binary

```python
def start_states(n):
    return np.random.randint(0, 2, n)
```

now map the start value to a sequence
could start with indices from arange
then a for loop with yield
then from iter
now could create an
or could make a zero array and then modify in place
could also make n choices from both distributions, convert to stacks, but nvmd

```python
def state_sequence(start, n):
    seq = np.zeros(n)
    seq[0] = start
    for i in range(n - 1):
        previous = seq[n]
        dist = rain_p[:, previous]
        seq[n + 1] = np.random.choice(2, p=dist)
    return seq
```
